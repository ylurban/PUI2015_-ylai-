```python
MSA_group = MSA_County_Area.groupby(' Name ')
```

```python
#Total Carbon Emission by MSA.
MSA_Total = MSA_group[' Total'].sum() 
MSA_Total.sort(ascending=False)
#Get land area, population and housing units by MSA.
MSA_Area = pd.merge(County_Area,df_MSA, on='FIPS', how = 'outer')
MSA_Area_group = MSA_Area.groupby('MSA Name ')
MSA_Total_Area = MSA_Area_group['Area in square miles - Land area'].sum()
MSA_Total_POP = MSA_Area_group['Population'].sum()/10000000  
MSA_Total_Housing = MSA_Area_group['Housing  units'].sum()/10000000 
MSA_Total_Area.sort(ascending=False)
MSA_Total_POP.sort(ascending=False)
MSA_Total_Housing.sort(ascending=False)
```

```python
#Land Area by MSA.
plt.ylabel('Land Area in Square Mile')
MSA_Total_Area[:10].plot(kind = 'bar', figsize = (10,6), title = 'Top 10 MSA with Largest Land Area in U.S. 2002',alpha = 0.7 )
```

```python
Carb_PerCap = MSA_Total/ MSA_Total_POP
Carb_PerCap.sort(ascending=False)
plt.ylabel('Household in Million')
Carb_PerCap.plot(kind = 'bar', color = 'blue' , figsize = (25,6), title = 'MSA with Carbon Emissions Per Capita in U.S. 2002', alpha = .8 )
#Carb_PerCap[398:340].plot(kind = 'bar', color = 'blue' , figsize = (25,6), title = 'MSA with Carbon Emissions Per Capita in U.S. 2002', alpha = .8 )
```

```python
MSA_SUM = pd.read_csv('MSA_Summary.csv')
```

```python
MSA_SUM['Density']=MSA_SUM['POP']/MSA_SUM['Land_Area']*10000000
MSA_SUM['Carb_PerCap'] = MSA_SUM['Total_Carb']/MSA_SUM['POP']*10000000
```

```python
MSA_SUM.corr()
```

```python
MSA_SUM[:10]
```

```python
mod = smf.ols(formula='Total_Carb ~ Density', data = MSA_SUM).fit()
print(mod.summary())
```

```python
mod = smf.ols(formula='Carb_PerCap ~ POP', data = MSA_SUM).fit()
print(mod.summary())
```

```python
#MSA_SUM['Total_Carb'].sort(ascending=False)
```

```python
x = log10(MSA_SUM['POP'])
y = log10(MSA_SUM['Carb_PerCap'])
model = sma.OLS(y,x) 
mod = model.fit()
mod.params[0]
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x,y,'k.', color='blue', alpha = .6)
plt.plot(x, mod.params[0]*x , '-', color = 'r')

plt.xlabel('Population (Million)', fontsize = 12)
plt.ylabel('Carbon Emission Per Capita (Ton)', fontsize = 12)
plt.title('Population Size vs. Total Carbon Emission', fontsize = 15)
```

```python
x = log10(MSA_SUM['Land_Area'])
y = log10(MSA_SUM['Total_Carb'])

model = sma.OLS(y, x) 
fit = model.fit()
fit.params[0]
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x,y, 'k.', color='blue', alpha = .6)
plt.plot(x, fit.params[0]*x , '-', color = 'r')


plt.xlabel('log-Population Density', fontsize = 12)
plt.ylabel('log-Carbon Emission Per Capita (Ton)', fontsize = 12)
plt.title('Population Density vs. Carbon Emission Per Capita', fontsize = 15)
```

```python
x = log10(MSA_SUM['POP'])
y = log10(MSA_SUM['Carb_PerCap'])

plt.figure(figsize=(8, 5))
plt.plot(x, y,'k.', color='blue', alpha = .6)

plt.xlabel('log- Population (Million)', fontsize = 12)
plt.ylabel('log- Carbon Emission per Capita (Ton)', fontsize = 12)
plt.title('Population Size vs. Carbon Emissions per Capita', fontsize = 15)
```
